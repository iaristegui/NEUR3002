```python

import numpy as np
import matplotlib.pyplot as plt

# General Parameters
n_conditions = 5 # number of conditions
n_spikes = 3 # number of spikes for the condition
duration = 100 # total time these spikes are in
n = 1 # starter variable for while loop

# Parameters pt. 2 - Synaptic weight factor (these determine how each stimulus impacts the firing rate)
conditioned_stimulus_weight = 1
unconditioned_stimulus_weight = 2.5
spike_times1 = []
spike_times2 = []
spike_times3 = []
spike_times4 = []
spike_times5 = []
for i in range(n_spikes): # generates 3 spikes for the baseline of each condition
    time1 = np.random.rand() * duration
    spike_times1.append(time1)
    time2 = np.random.rand() * duration
    spike_times2.append(time2)
    time3 = np.random.rand() * duration
    spike_times3.append(time3)
    time4 = np.random.rand() * duration
    spike_times4.append(time4)
    time5 = np.random.rand() * duration
    spike_times5.append(time5)
while n <= n_conditions:
    if n == 1: # generates a baseline condition
      for j in range(n_spikes):
        time1 = (np.random.rand() * duration) + duration
        spike_times1.append(time1)
    elif n == 2: # generates a Conditioned Stimulus only condition - firing rate will be the same as for baseline here
      for j in range(int(n_spikes*conditioned_stimulus_weight)):
        time2 = (np.random.rand() * duration) + duration
        spike_times2.append(time2)
    elif n == 3: # generates an Unconditioned Stimulus only condition
      for j in range(int(n_spikes*unconditioned_stimulus_weight)):
        time3 = (np.random.rand() * duration) + duration
        spike_times3.append(time3)
    elif n == 4: # generates a condition of the Conditioned and Unconditioned Stimulus at the same time
      n_spikes_CSUS = n_spikes*unconditioned_stimulus_weight*conditioned_stimulus_weight
      for j in range(int(n_spikes_CSUS)):
        time4 = (np.random.rand() * duration) + duration
        spike_times4.append(time4)
      if n_spikes_CSUS > 2: #simulate LTP by updating conditioned stimulus weight
        conditioned_stimulus_weight = conditioned_stimulus_weight*2.5
    else: # generates a condition of the conditioned stimulus after being matched with the Unconditioned Stimulus
      for j in range(int(n_spikes*conditioned_stimulus_weight)):
        time5 = (np.random.rand() * duration) + duration
        spike_times5.append(time5)
    n += 1
plt.eventplot([spike_times1, spike_times2, spike_times3, spike_times4, spike_times5], colors=['black'])
plt.xlabel('Time (ms)') # Label for x
plt.ylabel('Condition') # Label for y
plt.title('Raster Plots for a Postsynaptic Neuron in the Basal Amygdala Under Different Stimuli')
plt.yticks([0, 1, 2, 3, 4], ['Baseline', 'Conditioned Stimulus Alone', 'Unconditioned Stimulus Alone', 'Conditioned Stimulus + Unconditioned Stimulus', 'Conditioned Stimulus Post-Conditioning']) # labels the 5 Conditions on the Y Axis
plt.axvline(x=100, color='red', linestyle='-', linewidth=1, label = 'Stimulus') # labels the time when the stimulus is initiated
plt.show()

```
